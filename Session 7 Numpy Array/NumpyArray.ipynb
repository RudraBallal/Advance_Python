### Plot the curve x*sin(x)+x^2

```python
x = np.arange(-4*pi,4*pi,0.1)
x
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.title('Plot for y = x*np.sin(x) + x**2')
plt.show()
```
